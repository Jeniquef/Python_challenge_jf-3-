```python
import csv
import os

# Define file paths for budget and election data
file_to_load_budget = os.path.join("Resources", "budget_data.csv")
file_to_output_budget = os.path.join("analysis", "budget_analysis.txt")

file_to_load_election = os.path.join("Resources", "election_data.csv")
file_to_output_election = os.path.join("analysis", "election_analysis.txt") 

# Define variables to track the financial data
total_months = 0
net_total = 0
first_value = None
difference = []
months = []

# Variables for election
total_votes = 0
candidates = set()
charles_count = 0
Ray_count = 0
Diana_count = 0

# Process budget data
with open(file_to_load_budget, 'r') as csvfile:
    csvreader = csv.reader(csvfile)
    
    # Skip the header row
    header = next(csvreader)

    for row in csvreader:
        total_months += 1  # Count months per row 
        
        month = row[0]  # Get the current month
        pl = int(row[1])  # Profit/loss value

        months.append(month)  # Add the current month to the list of months 
        net_total += pl  # Sum of profits/losses
        
        if first_value is not None: 
            change = pl - first_value  # Get the difference between the current profit/loss and the previous one
            difference.append(change)  # Add the difference to the change list
            
        first_value = pl  # Set the current value as the first_value for the next iteration 

# Get the average difference between the months 

    ad = sum(difference) / len(difference)  # Average change based on the number of changes
    formatted_amount = f"${ad:,.2f}"


# Find the max and min difference 

    max_difference = max(difference) 
    min_difference = min(difference)

    max_index = difference.index(max_difference) + 1  # Offset by 1 to pick the month after the difference 
    min_index = difference.index(min_difference) + 1  

# store results ( budget analysis)
output_budget = (
    f"Total Months: {total_months}\n"
    f"Net Total: ${net_total:,.2f}\n"
    f"Average Change: {formatted_amount}\n"
    f"Greatest Increase in Profits: {months[max_index]} (${max_difference})\n"
    f"Greatest Decrease in Profits: {months[min_index]} (${min_difference})\n"
)

# Write the budget results output to a text file
with open(file_to_output_budget, 'w') as txt_file:
    txt_file.write(output_budget)

print("budget results  exported to budget_analysis.txt")

# Process election data
with open(file_to_load_election, 'r') as csvfile:
    csvreader = csv.reader(csvfile)
    
    # Skip the header
    next(csvreader)

    for row in csvreader:
        total_votes += 1 
        candidates.add(row[2]) 
        
        # Count votes for each candidate
        if row[2] == "Charles Casper Stockham":  
            charles_count += 1  
        elif row[2] == "Raymon Anthony Doane": 
            Ray_count += 1 
        elif row[2] == "Diana DeGette":  
            Diana_count += 1  

# Calculate percentages
cvp = (charles_count / total_votes) * 100 
rvp = (Ray_count / total_votes) * 100 
dvp = (Diana_count / total_votes) * 100 


#max votes 
max_votes = max(charles_count, Ray_count, Diana_count)
if max_votes == charles_count:
    winner = "Charles Casper Stockham"
elif max_votes == Ray_count:
    winner = "Raymon Anthony Doane"
else:
    winner = "Diana DeGette"

# store for election results
output_election = (
    "Election Results\n"
    "---------------------------------\n"
    f"Total Votes: {total_votes}\n"
    "---------------------------------\n"
    f"Charles Casper Stockham: {charles_count} ({cvp:.2f}%)\n"
    f"Raymon Anthony Doane: {Ray_count} ({rvp:.2f}%)\n"
    f"Diana DeGette: {Diana_count} ({dvp:.2f}%)\n"
    "---------------------------------\n"
    f"Winner: {winner}\n"
)



# Write the election results output to a text file
with open(file_to_output_election, 'w') as txt_file:
    txt_file.write(output_election)

print("Election results have been exported to election_analysis.txt")
```
