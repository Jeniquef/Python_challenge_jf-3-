```python
import csv

total_votes = 0 
candidates = set()  # Use 'candidates' instead of 'canadiates'
charles_count = 0 
Ray_count = 0 
Diana_count = 0 
cvp = 0 
rvp = 0 
dvp = 0 

# Set the path for the CSV file
p = 'election_data.csv'  

with open(p, 'r') as csvfile:
    csvreader = csv.reader(csvfile)
    
    # Skip the header
    next(csvreader)

    for row in csvreader:
        total_votes += 1  # Increment total votes for each row
        candidates.add(row[2])  # Add candidate name to the set
        
        # Count votes for Charles
        if row[2] == "Charles Casper Stockham":  # Check the candidate name
            charles_count += 1  # Increment the counter if the candidate is Charles
    
        if row[2] == "Raymon Anthony Doane":  # Check the candidate name
            Ray_count += 1  # Increment the counter if the candidate is Charles
             
        
        if row[2] == "Diana DeGette":  # Check the candidate name
            Diana_count += 1  # Increment the counter if the candidate is Charles



cvp = charles_count/total_votes*100
(cvp)

rvp = Ray_count/total_votes*100
(rvp)

dvp = Diana_count/total_votes*100
(dvp)



# Convert candidates set to a list
candidates_list = list(candidates)

# Print results

print(f"Election Results")

print(f"---------------------------------")
print(f"Total Votes: {total_votes}")

print(f"---------------------------------")

print(f"Charles Casper Stockham: {cvp:.2f}%",charles_count)

print(f"Raymon Anthony Doane: {rvp:.2f}%", Ray_count)

print(f"Diana DeGette: {dvp:.2f}%", Diana_count)

print(f"---------------------------------")

max_votes = max(charles_count, Ray_count, Diana_count)

if max_votes == charles_count:
    print("Winner :Charles Casper Stockham ")
elif max_votes == Ray_count:
    print("Winner :Raymon Anthony Doane ")
else:
    print("Winner : Diana DeGette ")

#print(f"Candidates: {candidates_list}")
#print(f"Charles Casper Stockham count: {charles_count}")
#print(f"Raymon Anthony Doane count: {Ray_count}")
#print(f"Diana DeGette: {Diana_count}")
```
